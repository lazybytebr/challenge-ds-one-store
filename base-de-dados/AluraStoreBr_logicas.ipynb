### Importação dos dados



```python
import pandas as pd

url  = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja  = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja4.head()
```

#1. Análise do faturamento


```python
# listar todas as colunas do dataset objeto
print( loja.columns )

# colunas em lormato de lista
print( loja.columns.tolist() )

# Comparar se todas as colunas são iguais (ignorando ordem)
colunas_iguais = ( set( loja.columns ) == set( loja2.columns ) == set( loja3.columns ) == set( loja4.columns ) )
print("Todas as colunas são iguais", colunas_iguais)
```

```python
# Somando os valores da coluna Preço de cada loja para estimar o faturamento de cada loja.
soma_loja = loja['Preço'].sum()
soma_loja2 = loja2['Preço'].sum()
soma_loja3 = loja3['Preço'].sum()
soma_loja4 = loja4['Preço'].sum()

print( "Faturamento Loja  R$", soma_loja, "\nFaturamento Loja2 R$", soma_loja2, "\nFaturamento Loja3 R$", soma_loja3, "\nFaturamento Loja4 R$", soma_loja4 )
```

```python
# juntando os valores
faturamento = {"loja1":soma_loja, "loja2":soma_loja2, "loja3":soma_loja3, "loja4":soma_loja4}

# adicionando a um DataFrame
faturamentos = pd.DataFrame(list(faturamento.items()), columns=["Loja", "Faturamento"])

# ordenando o DataFrame
faturamentos_ordenados = faturamentos.sort_values( by='Faturamento', ascending= False)

print(faturamentos_ordenados)
```

# 2. Vendas por Categoria


```python
# Agrupar os dados por categoria e contar o número de vendas mostrando as categoria mais populares de cada loja
contagem_loja1 = loja.groupby('Categoria do Produto')['Produto'].count().reset_index(name='total_vendas')
contagem_loja2 = loja2.groupby('Categoria do Produto')['Produto'].count().reset_index(name='total_vendas')
contagem_loja3 = loja3.groupby('Categoria do Produto')['Produto'].count().reset_index(name='total_vendas')
contagem_loja4 = loja4.groupby('Categoria do Produto')['Produto'].count().reset_index(name='total_vendas')


contagem_loja1 = contagem_loja1.sort_values(by='total_vendas', ascending = False)
contagem_loja2 = contagem_loja2.sort_values(by='total_vendas', ascending = False)
contagem_loja3 = contagem_loja3.sort_values(by='total_vendas', ascending = False)
contagem_loja4 = contagem_loja4.sort_values(by='total_vendas', ascending = False)

print('Loja 01\n',contagem_loja1, '\n')
print('Loja 02\n',contagem_loja2, '\n')
print('Loja 03\n',contagem_loja3, '\n')
print('Loja 04\n',contagem_loja4, '\n')
```

# 3. Média de Avaliação das Lojas

```python
# Entender a satisfação dos clientes com os produtos vendidos.
satisfacao_loja1 = round(loja['Avaliação da compra'].sum() / len(loja['Avaliação da compra']),2)
satisfacao_loja2 = round(loja2['Avaliação da compra'].sum() / len(loja2['Avaliação da compra']),2)
satisfacao_loja3 = round(loja3['Avaliação da compra'].sum() / len(loja3['Avaliação da compra']),2)
satisfacao_loja4 = round(loja4['Avaliação da compra'].sum() / len(loja4['Avaliação da compra']),2)

print('A média das avaliações da loja 01 foi de', satisfacao_loja1,'pontos\n' )
print('A média das avaliações da loja 02 foi de', satisfacao_loja2,'pontos\n' )
print('A média das avaliações da loja 03 foi de', satisfacao_loja3,'pontos\n' )
print('A média das avaliações da loja 04 foi de', satisfacao_loja4,'pontos\n' )
```

# 4. Produtos Mais e Menos Vendidos

```python
# Identificar quais produtos foram os mais vendidos e os menos vendidos em cada loja
print(f"Na Loja 01 o a categoria menos vendida é {contagem_loja1['Categoria do Produto'][0]} com {contagem_loja1['total_vendas'][0]} vendas, e a com menos vendas é {contagem_loja1['Categoria do Produto'].iloc[-1]} com {contagem_loja1['total_vendas'].iloc[-1]} vendas!\n ")

print(f"Na Loja 02 o a categoria menos vendida é {contagem_loja2['Categoria do Produto'][0]} com {contagem_loja2['total_vendas'][0]} vendas, e a com menos vendas é {contagem_loja2['Categoria do Produto'].iloc[-1]} com {contagem_loja2['total_vendas'].iloc[-1]} vendas!\n ")

print(f"Na Loja 03 o a categoria menos vendida é {contagem_loja3['Categoria do Produto'][0]} com {contagem_loja3['total_vendas'][0]} vendas, e a com menos vendas é {contagem_loja3['Categoria do Produto'].iloc[-1]} com {contagem_loja3['total_vendas'].iloc[-1]} vendas!\n ")

print(f"Na Loja 04 o a categoria menos vendida é {contagem_loja4['Categoria do Produto'][0]} com {contagem_loja4['total_vendas'][0]} vendas, e a com menos vendas é {contagem_loja4['Categoria do Produto'].iloc[-1]} com {contagem_loja4['total_vendas'].iloc[-1]} vendas!\n ")
```

# 5. Frete Médio por Loja


```python
# Entender quanto, em média, está sendo gasto com frete para cada uma das lojas.
frete_medio_1 = round( loja['Frete'].sum() / len(loja['Frete']), 2 )
frete_medio_2 = round( loja2['Frete'].sum() / len(loja2['Frete']), 2 )
frete_medio_3 = round( loja3['Frete'].sum() / len(loja3['Frete']), 2 )
frete_medio_4 = round( loja4['Frete'].sum() / len(loja4['Frete']), 2 )

print(f"O frete médio da loja 01 é de {frete_medio_1}, o da loja 02 é {frete_medio_2}, o da loja 03 é {frete_medio_3} e o da loja04 é {frete_medio_4}")
```

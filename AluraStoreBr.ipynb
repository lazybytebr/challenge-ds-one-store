# Análise de dados para a Store

## Introdução

O documento abaixo é uma análise exploratória dos dados das lojas do cliente Store, tendo o objetivo de facilitar a visualização do desempenho e comportamentos de cada filial a fim de facilitar a escolha sobre qual loja deverá ser vendida pelo proprietário.

### Importação dos dados



```python
import pandas as pd

url  = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja  = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Somando os valores da coluna Preço de cada loja para estimar o faturamento de cada loja.
soma_loja = loja['Preço'].sum()
soma_loja2 = loja2['Preço'].sum()
soma_loja3 = loja3['Preço'].sum()
soma_loja4 = loja4['Preço'].sum()

print( f"Faturamento Total da Loja1 R$ { soma_loja } \nFaturamento Total da Loja2 R$ { soma_loja2 } \nFaturamento Total da Loja3 R$ { soma_loja3 } \nFaturamento Total da Loja4 R$ { soma_loja4 }" )
```

```python
todas_lojas = soma_loja + soma_loja2 + soma_loja3 + soma_loja4
print(f"Faturamento Total das Lojas: R$ {todas_lojas}")
```

```python
import matplotlib.pyplot as plt

# Criar um dicionário com os dados
faturamento_lojas = {
    'Loja 1': soma_loja,
    'Loja 2': soma_loja2,
    'Loja 3': soma_loja3,
    'Loja 4': soma_loja4,
}

# Criar o gráfico de barras
plt.figure(figsize=(15, 8))
bars = plt.bar(faturamento_lojas.keys(), faturamento_lojas.values(), color='lightblue', edgecolor='black', linewidth=1)
plt.xticks(rotation=30)

# Adicionar título e rótulos
plt.gca().yaxis.set_major_formatter('R${x:.2f}')
plt.title('Faturamento Total por Loja (filial)')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Objeto de eixo
ax = plt.gca()

# Adicionar rótulos de dados em cima de cada barra
ax.bar_label(bars, labels=[f'R${valor:,.2f}' for valor in faturamento_lojas.values()], padding=3)

plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
#criaçao de um dicionario para armazenar os dados das 4 lojas
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

for nome_loja, df in lojas.items():
    print(f'\nCategorias mais vendidas - {nome_loja}')
    categorias = df['Categoria do Produto'].value_counts()
    total_vendas = categorias.values.sum()
    print(f"Total de itens vendidos em todas as categorias: {loja} {total_vendas} ")
    print(categorias)

# Criando subplots
fig, axs = plt.subplots(2, 2, figsize=(18, 10))
fig.suptitle('Categorias Mais Vendidas por Loja', fontsize=14)

# Loop para plotar cada gráfico
for i, (nome_loja, df) in enumerate(lojas.items()):
    categorias = df['Categoria do Produto'].value_counts()

    ax = axs[i // 2, i % 2]
    bars = ax.bar(categorias.index, categorias.values, color='lightgreen')

    ax.set_title(nome_loja)
    ax.set_ylabel('Quantidade Vendida')
    ax.set_xlabel('Categoria')
    ax.tick_params(axis='x', rotation=30)

    # Adiciona rótulos de dados em cima de cada barra
    ax.bar_label(bars, labels=categorias.values, padding=3)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 3. Média de Avaliação das Lojas

```python
# Entender a satisfação dos clientes com os produtos vendidos.
satisfacao_loja1 = round(loja['Avaliação da compra'].sum() / len(loja['Avaliação da compra']),2)
satisfacao_loja2 = round(loja2['Avaliação da compra'].sum() / len(loja2['Avaliação da compra']),2)
satisfacao_loja3 = round(loja3['Avaliação da compra'].sum() / len(loja3['Avaliação da compra']),2)
satisfacao_loja4 = round(loja4['Avaliação da compra'].sum() / len(loja4['Avaliação da compra']),2)

print('A média das avaliações da loja 01 foi de', satisfacao_loja1,'pontos\n' )
print('A média das avaliações da loja 02 foi de', satisfacao_loja2,'pontos\n' )
print('A média das avaliações da loja 03 foi de', satisfacao_loja3,'pontos\n' )
print('A média das avaliações da loja 04 foi de', satisfacao_loja4,'pontos\n' )
```

```python
lojas = ["Loja 01", "Loja 02", "Loja 03", "Loja04" ]
medias_aval = [satisfacao_loja1,satisfacao_loja2,satisfacao_loja3,satisfacao_loja4]

# Criando o gráfico
plt.figure(figsize=(8, 5))
plt.plot(lojas, medias_aval, marker='o', linestyle='-', color='orange', linewidth=2)

# Adicionando os valores acima de cada ponto
for i, valor in enumerate(medias_aval):
    plt.text(i, valor + 0.1, f'{valor:.2f}', ha='center', va='bottom', fontsize=10)

# Título e rótulos
plt.title('Média de Avaliações por Loja', fontsize=15)
plt.xlabel('Lojas')
plt.ylabel('Média de Avaliação')
plt.ylim(0, 5)  # Supondo que a avaliação vai de 0 a 5

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
import matplotlib.pyplot as plt


# Criar um dicionario para todas as lojas
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

#repetição para fazer a leitura de valores de cada loja mais vendido
for nome_loja, df in lojas.items():
    # Encontrar o produto mais vendido para cada loja
    maior_venda = df['Produto'].value_counts().idxmax()
    quantidade_vendida = df['Produto'].value_counts().max()
    print(f"O Produto mais vendido da {nome_loja} é: {maior_venda} com {quantidade_vendida} vendas")
```

```python
#Gerar o Grafico Produtos mais vendidos por loja

import matplotlib.pyplot as plt

# Criar listas para armazenar os dados do loop
nomes_lojas = []
quantidades = []
produtos = []

# Reexecutar o loop apenas para armazenar os dados
for nome_loja, df in lojas.items():
    maior_venda = df['Produto'].value_counts().idxmax()
    quantidade_vendida = df['Produto'].value_counts().max()

#Para adicionar um elemento ao final da lista e exibir as 4 lojas no grafico
    nomes_lojas.append(nome_loja)
    quantidades.append(quantidade_vendida)
    produtos.append(maior_venda)

# Criar o gráfico de barras
plt.figure(figsize=(12, 6))
barras = plt.bar(nomes_lojas, quantidades, color='lightblue')

# Adicionar rótulos nas barras (produto + quantidade)
for i, barra in enumerate(barras):
    altura = barra.get_height()
    rotulo = f"{produtos[i]}\n({quantidades[i]})"
    plt.text(barra.get_x() + barra.get_width()/2, altura + 1, rotulo, ha='center', va='bottom', fontsize=10, fontweight='bold')

# Títulos e ajustes
plt.title('Produto Mais Vendidos por Loja')
plt.xlabel('Lojas')
plt.ylabel('Quantidade Vendida')
plt.ylim(0, max(quantidades) + 10)
plt.tight_layout()
plt.show()
```

```python
#repetição para fazer a leitura de valores de cada loja menos vendido
for nome_loja, df in lojas.items():
    # Encontrar o produto menos vendido para cada loja
    menor_venda = df['Produto'].value_counts().idxmin()
    print(f"O Produto menos vendido da {nome_loja} é: {menor_venda} com {quantidade_vendida} vendas")
```
